# Step4. DATA  PRE-PROCESSING

### 4.0 Handling high correlation

```python
correlation_matrix = loan[numerical_columns].corr().abs()
highly_correlated_pairs = set()

for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if correlation_matrix.iloc[i, j] > 0.8:
            colname = correlation_matrix.columns[i]
            highly_correlated_pairs.add(colname)

print(f"Highly correlated features: {highly_correlated_pairs}")

#Drop one feature from each highly correlated pair
loan_dropped = loan.drop(columns=highly_correlated_pairs)

#Update the list of numerical columns
updated_numerical_columns = [col for col in numerical_columns if col not in highly_correlated_pairs]

print(f"Features dropped: {highly_correlated_pairs}")

#Check the updated correlation matrix
updated_correlation_matrix = loan_dropped[updated_numerical_columns].corr()
print(updated_correlation_matrix)

```

### 4.1 Handling Categorical data

Converting person_gender, person_education(ordinal), person_home_ownership, loan_intent, and previous_loan_defaults_on_file  to numerical values using encoding techniques.(OneHotEncoding and label coding for nominal and ordinal categories respectifully)

```python
label_encoder = LabelEncoder()

# Encode binary categorical columns (person_gender, previous_loan_defaults_on_file)
loan['person_gender'] = label_encoder.fit_transform(loan['person_gender'])
loan['previous_loan_defaults_on_file'] = label_encoder.fit_transform(loan['previous_loan_defaults_on_file'])

# One-hot encoding for other categorical columns
loan = pd.get_dummies(loan, columns=['person_education', 'person_home_ownership', 'loan_intent'], drop_first=True)

# Verify if they have been encoded accordingly
loan.head()



```

```python
#checking the columns after encoding the data
loan.columns
```

### 4.1 Feature scaling
standardizing / normalizing the features to ensure they all contribute to the model

```python

numerical_features = ['person_income', 'person_emp_exp', 'loan_amnt', 'loan_int_rate', 'loan_percent_income', 
                      'cb_person_cred_hist_length', 'credit_score']

# Initialize StandardScaler and scale the features
scaler = StandardScaler()
loan[numerical_features] = scaler.fit_transform(loan[numerical_features])

# Verify the scaling
loan.head()


```

# Step5. BUILDING THE INITIAL BASELINE MODEL

### Step 5.1 Logistic Regression Model

```python


# Initialize and train the logistic regression model
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)

# Make predictions
y_pred_log_reg = log_reg.predict(X_test)
# Get predicted probabilities for the positive class
y_pred_proba_log_reg = log_reg.predict_proba(X_test)[:, 1]

# Computing the ROC-AUC score
roc_auc = roc_auc_score(y_test, y_pred_proba_log_reg)  
# Evaluate performance
print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred_log_reg))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_log_reg))
print("Classification Report:\n", classification_report(y_test, y_pred_log_reg))

print("ROC-AUC Score:", roc_auc)


```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba_log_reg)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

### Interpreting the perfomance metrics

1. ROC-AUC score is 0.95

Indicates that the logistic regression model is performing well. It is doing a good job of distinguishing between the positive and negative classes.

2. Accuracy 0.8938
Indicates that the model correctly predicted the class (either 0 or 1) for 89.38% of all instances in the dataset.So the overall perfomance of the model is good

3. Confusion Matrix

# Step6. Use a Decision Tree Classifier

From the perfomance metrics above,our data has shown a class imbalance...Decision Trees can handle class imbalance well and allow for deeper splits that might help identify loan approvals(class 1) more effectively.Adjust class_weight parameter to give more weight to class 1, prioritizing loan approvals)

```python

dtmodel = DecisionTreeClassifier(class_weight={0: 1, 1: 2}, random_state=42)
dtmodel.fit(X_train, y_train)

# Make predictions
y_pred_dt = dtmodel.predict(X_test)

# Get predicted probabilities 
y_pred_proba_dt = dtmodel.predict_proba(X_test)[:, 1]

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba_dt)

# Calculate AUC (Area Under the Curve)
roc_auc = auc(fpr, tpr)

# Evaluate performance

print("ROC-AUC:\n", roc_auc)

print("Decision Tree Accuracy:", accuracy_score(y_test, y_pred_dt))

print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_dt))

print("Classification Report:\n", classification_report(y_test, y_pred_dt))
```

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba_dt)
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

# Step8. KNN(KNeighborsClassifier) Classifier

Since KNN uses distance metrics and can be effective in capturing complex relationships, it's worth trying with a weighted approach (using 'distance' weights or class weights) to improve recall for class 1.

```python
# Initialize KNN with distance-based weighting
knn = KNeighborsClassifier(weights='distance')
knn.fit(X_train, y_train)

# Make predictions
y_pred_knn = knn.predict(X_test)

# Get predicted probabilities 
y_pred_proba_knn = knn.predict_proba(X_test)[:, 1]

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba_knn)

# Calculate AUC (Area Under the Curve)
roc_auc = auc(fpr, tpr)

# Evaluate performance

print("ROC-AUC:\n", roc_auc)

# Evaluate performance
print("KNN Accuracy:", accuracy_score(y_test, y_pred_knn))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
print("Classification Report:\n", classification_report(y_test, y_pred_knn))
```

### Interpreting KNN metrics in comparison to Random Forest
The K-Nearest Neighbors (KNN) model achieved an accuracy of 89.71%, which is slightly lower than the Random Forest model (92.81%)

The recall for class 1 (loan approved) is 70%, which is lower than the Random Forest recall of 77%. This means that KNN is missing more actual loan approval cases compared to Random Forest.

Precision for class 1 is 81%, which is lower than Random Forest's precision of 90%, meaning that the KNN model is misclassifying more loan denials as loan approvals.

The recall for class 0 (loan not approved) is quite good at 95%, which means the model is quite strong at identifying loan denials.

IN conclusion;

1. Random Forest remains the best-performing model overall, achieving the highest accuracy, F1-score, ROC-AUC score, and precision for class 1.

2. KNN performs decently but lags behind Random Forest and Decision Tree in several areas, including recall for class 1 and F1-score for class 1. However, it is still competitive with a ROC-AUC score of 0.9227.

3. Decision Tree and Logistic Regression are more balanced, with Decision Tree slightly outperforming Logistic Regression in terms of F1-score and recall for class 1.

### Tuning the hyper parameters for the KNN model and evaluating the perfomance again

```python
# Define hyperparameter grid for KNN
param_grid = {
    'n_neighbors': [3, 5, 7, 10, 15],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

# Initialize GridSearchCV for hyperparameter tuning
#applying a cross validation of cv=5 to prevent overfitting
grid_search_knn = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring='recall')
grid_search_knn.fit(X_train, y_train)

# Best hyperparameters
print("Best Hyperparameters:", grid_search_knn.best_params_)

# Retrain with best parameters
tuned_knn = grid_search_knn.best_estimator_
y_pred_tuned_knn = tuned_knn.predict(X_test) #getting the predictions

# Get predicted probabilities 
y_pred_proba_tuned_knn = tuned_knn.predict_proba(X_test)[:, 1]

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba_tuned_knn)

# Calculate AUC (Area Under the Curve)
roc_auc = auc(fpr, tpr)

# Evaluate performance

print("ROC-AUC:\n", roc_auc)
print("Tuned KNN Accuracy:", accuracy_score(y_test, y_pred_tuned_knn))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_tuned_knn))
print("Classification Report:\n", classification_report(y_test, y_pred_tuned_knn))
```

### Visualizing the ROC-AUC curve and the confusion matrix

```python
# Set up the figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot ROC curve
axes[0].plot(fpr, tpr, label=f'ROC Curve (AUC = {roc_auc:.2f})')
axes[0].plot([0, 1], [0, 1], 'k--')  
axes[0].set_xlabel('False Positive Rate')
axes[0].set_ylabel('True Positive Rate')
axes[0].set_title('ROC Curve')
axes[0].legend(loc='upper right')

# Plot Confusion Matrix
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='coolwarm', xticklabels=['Predicted Negative', 'Predicted Positive'], yticklabels=['Actual Negative', 'Actual Positive'], ax=axes[1])
axes[1].set_xlabel('Predicted')
axes[1].set_ylabel('Actual')
axes[1].set_title('Confusion Matrix')
plt.tight_layout()
plt.show()
```
